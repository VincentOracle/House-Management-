```python
                # QUESTION1 

# Using the pandas package, this code first imports the training and 
# testing data into dataframes. Following that, it separates the training and 
# test sets of data into features (X) and labels (Y).
# It then initializes a logistic regression model and applies the fit method to train it using the training data.
# The accuracy_score function from the metrics module of scikit-learn
# is then used to assess the performance of the trained model by making
# predictions on the test data. 
# The percentage of correctly categorized samples in the test data is represented by the accuracy score.


# Split the data into features and labels
X_train = train_data.drop("y", axis=1)
y_train = train_data["y"]
X_test = test_data.drop("y", axis=1)
y_test = test_data["y"]
```

```python
# Train the logistic regression model
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
  # QUESTION 2 
# To make a prediction for a new instance, we first need to compute the linear 
# combination of its features and the model's weights:
# z = w0 + i wiXi
# We then apply the sigmoid function to z to obtain the predicted probability of the positive class:
# p(y=1|X) = 1 / (1 + exp(-z))
# Finally, we can apply a threshold of 0.5 to this probability to obtain the predicted class label:
# y_pred = 1 if p(y=1|X) >= 0.5 else -1
# In Python, we can implement this prediction algorithm as follows, assuming that we 
# have already trained a logistic regression model and loaded the new instance X as a numpy array
```

```python
import numpy as np
import pandas as pd

# Load the testing data from TestData.csv
test_data = pd.read_csv("C:\\Users\\n\\Downloads\\ML PROJECT\\TestData (1).csv")

# Extract the feature values from the testing data
X = test_data.iloc[:, :-1].values

# Make predictions using the logistic regression model
# Compute the linear combination of features and weights
z = np.dot(X, lr.coef_.T) + lr.intercept_
# Apply the sigmoid function to obtain the predicted probability
prob = 1 / (1 + np.exp(-z))
# Apply the threshold to obtain the predicted class label
y_pred = np.where(prob >= 0.5, 1, -1)
print(y_pred)
```

```python
         #Evaluating classifier accuracy
# To evaluate the accuracy of the logistic regression classifier on the test data, the following code is has been used.
# Here, we first load the testing data into a pandas DataFrame using pd.read_csv, and then extract the feature values X_test 
# and true class labels y_test from the DataFrame. We then make predictions using the logistic regression model,
# as described in the previous answer. Finally, we compute the accuracy of the classifier by comparing the
# predicted class labels y_pred to the true class labels y_test, and counting the number of correct predictions 
# divided by the total number of instances in the testing set. The result is printed to the console using f-strings.
```

```python
import pandas as pd
# Load the testing data from TestData.csv

test_data = pd.read_csv("C:\\Users\\n\\Downloads\\ML PROJECT\\TestData (1).csv")
# Extract the feature values and true class labels from the testing data
X_test = test_data.iloc[:, :-1].values
y_test = test_data.iloc[:, -1].values
```

```python
# Make predictions using the logistic regression model
z = np.dot(X_test, lr.coef_.T) + lr.intercept_
prob = 1 / (1 + np.exp(-z))
y_pred = np.where(prob >= 0.5, 1, -1)
print(y_pred)
```

```python
# Compute the accuracy of the classifier
accuracy = sum(y_pred == y_test) / len(y_test)
print(f"Accuracy: {accuracy}")
```
